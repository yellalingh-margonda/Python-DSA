# ðŸ‘‘ N-Queens

## FLAG â€“ Is at least one solution possible?

```
def is_valid(row, col, board, n):
    for r in range(row):
        if board[r][col] == 'Q': return False
    for r,c in zip(range(row-1,-1,-1), range(col-1,-1,-1)):
        if board[r][c]=='Q': return False
    for r,c in zip(range(row-1,-1,-1), range(col+1,n)):
        if board[r][c]=='Q': return False
    return True

def nqueens_flag(n, row=0, board=None):
    if board is None:
        board = [['.' for _ in range(n)] for _ in range(n)]
    if row == n:
        return True
    for col in range(n):
        if is_valid(row, col, board, n):
            board[row][col] = 'Q'
            if nqueens_flag(n, row+1, board):
                return True
            board[row][col] = '.'
    return False
```

## ALL â€“ Return all solutions

```
def nqueens_all(n, row=0, board=None, res=None):
    if board is None:
        board = [['.' for _ in range(n)] for _ in range(n)]
    if res is None:
        res = []
    if row == n:
        res.append([''.join(r) for r in board])
        return
    for col in range(n):
        if is_valid(row, col, board, n):
            board[row][col] = 'Q'
            nqueens_all(n, row+1, board, res)
            board[row][col] = '.'
    return res
```

## COUNT â€“ Number of solutions

```
def nqueens_count(n, row=0, board=None):
    if board is None:
        board = [['.' for _ in range(n)] for _ in range(n)]
    if row == n:
        return 1
    count = 0
    for col in range(n):
        if is_valid(row, col, board, n):
            board[row][col] = 'Q'
            count += nqueens_count(n, row+1, board)
            board[row][col] = '.'
    return count
```

# ðŸ§© Sudoku Solver

## Helper Functions

```
def find_empty(board):
    for i in range(9):
        for j in range(9):
            if board[i][j] == 0:
                return i, j
    return None

def valid(board, r, c, num):
    if num in board[r]: return False
    if num in [board[i][c] for i in range(9)]: return False
    br, bc = 3*(r//3), 3*(c//3)
    for i in range(br, br+3):
        for j in range(bc, bc+3):
            if board[i][j] == num:
                return False
    return True
```

## FLAG â€“ Solve once

```
def sudoku_flag(board):
    pos = find_empty(board)
    if not pos:
        return True
    r, c = pos
    for num in range(1,10):
        if valid(board, r, c, num):
            board[r][c] = num
            if sudoku_flag(board):
                return True
            board[r][c] = 0
    return False
```

## COUNT â€“ Number of valid solutions

```
def sudoku_count(board):
    pos = find_empty(board)
    if not pos:
        return 1
    r, c = pos
    count = 0
    for num in range(1,10):
        if valid(board, r, c, num):
            board[r][c] = num
            count += sudoku_count(board)
            board[r][c] = 0
    return count
```
