This notebook explains the **thought process, template, code, and complexity** for permutation problems.

### Golden Question
**Who goes in the current position?**

---

## ðŸ§  Why permutation is different
- Subsequence â†’ pick / not pick
- Combination â†’ choose next option
- **Permutation â†’ place every element in every position**

Each level fixes **one position**.

---

## ðŸ§© Mental Template
```
if base case:
    save permutation

for each unused choice:
    choose
    recurse
    undo
```

---

````
# Method 1: Used-array method (most common)
def permute(nums):
    res = []
    used = [False] * len(nums)

    def backtrack(path):
        if len(path) == len(nums):
            res.append(path[:])
            return

        for i in range(len(nums)):
            if used[i]:
                continue
            used[i] = True
            path.append(nums[i])
            backtrack(path)
            path.pop()
            used[i] = False

    backtrack([])
    return res

print(permute([1,2,3]))
````

````
# Method 2: Swap-based method (space optimized)
def permute_swap(nums):
    res = []

    def backtrack(index):
        if index == len(nums):
            res.append(nums[:])
            return

        for i in range(index, len(nums)):
            nums[index], nums[i] = nums[i], nums[index]
            backtrack(index + 1)
            nums[index], nums[i] = nums[i], nums[index]

    backtrack(0)
    return res

print(permute_swap([1,2,3]))
````

## â± Time & Space Complexity
- Time: **O(n Ã— n!)**
- Space: **O(n)** recursion depth

---

## âŒ Why pick / not pick does NOT work
- Pick / not pick creates subsets
- Permutations require **ordering**
- Every element must appear in **every position**
