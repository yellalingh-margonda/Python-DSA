## ðŸ” Common Backtracking Skeleton

## 1ï¸âƒ£ Template: Check if ONE solution exists (Boolean / Flag)

````
def exists_solution(level):
    if level == N:
        return True  # Found one valid solution

    for choice in choices:
        if valid(choice):
            make(choice)
            if exists_solution(level + 1):
                return True  # EARLY EXIT
            undo(choice)

    return False
````

### ðŸ§  When to use
- Is it possible?
- Check solvability (e.g., N-Queens possible?)
- Find ANY valid answer


## 2ï¸âƒ£ Template: Find ALL solutions (Collect paths)

````
def all_solutions(level):
    if level == N:
        save_solution()
        return

    for choice in choices:
        if valid(choice):
            make(choice)
            all_solutions(level + 1)
            undo(choice)
````

### ðŸ§  When to use
- Print all answers
- Return all configurations
- Problems asking for "all ways"


## 3ï¸âƒ£ Template: COUNT number of solutions

````
def count_solutions(level):
    if level == N:
        return 1  # One valid configuration

    count = 0
    for choice in choices:
        if valid(choice):
            make(choice)
            count += count_solutions(level + 1)
            undo(choice)

    return count
````

### ðŸ§  When to use
- How many ways?
- Count subsets / subsequences / arrangements
- Do NOT store paths (just count)
