
# Recursion & Backtracking – Master Notes

These notes are designed so that **whenever you see recursion in the future,
you can instantly recognize backtracking**.



## Golden Rule (Very Important)

> **If the work (print / calculation) is done _after_ the recursive call, it is called BACKTRACKING.**

Memory trigger:
```
Before recursive call  → Normal recursion
After recursive call   → Backtracking
```



## Example 1: Print 1 to n (Backtracking)


````

def print_1_n(i):
    if i <= 0:
        return
    print_1_n(i - 1)
    print(i)

print_1_n(5)
````


### Why this is Backtracking?
- Recursive call happens first
- Print happens while returning back through the call stack



## Example 2: Print n to 1 (Normal Recursion)


````

def print_n_1(i):
    if i <= 0:
        return
    print(i)
    print_n_1(i - 1)

print_n_1(5)
````


### Why this is NOT Backtracking?
- Print happens before recursive call



## Example 3: Forward recursion using two variables


````

def print2(i, n):
    if i > n:
        return
    print(i)
    print2(i + 1, n)

print2(1, 5)
````


### Key Point
- Work is done before recursion
- This is not backtracking



## Example 4: Backtracking using two variables


````

def print3(i, n):
    if i > n:
        return
    print3(i + 1, n)
    print(i)

print3(1, 5)
````


### Why this IS Backtracking?
- Recursive call first
- Print happens during stack unwinding



## One‑Sentence Definition of Backtracking

> **Backtracking is when the recursive call is made first and the actual work is done while returning back through the call stack.**



## Final Mental Checklist

Ask yourself:
1. Is the work done after the recursive call? → ✅ Backtracking
2. Is the work done before the recursive call? → ❌ Not backtracking
