# Subsequence Problem (Recursion & Backtracking)

This notebook explains the **Subsequence problem**, its intuition, and solution. It also lists other problems where the **same recursion logic applies**.


## ðŸ“Œ Problem Statement

Given a string `s`, generate **all possible subsequences** of the string.

**Definition:** A subsequence is formed by deleting zero or more characters from the string without changing the relative order of the remaining characters.

**Example:**
Input: `"abc"`

Output:
`['abc', 'ab', 'ac', 'a', 'bc', 'b', 'c', '']`


## ðŸ§  Key Insight

For every character in the string, we have **two choices**:

1. Include the character in the current subsequence
2. Exclude the character

This leads to a **binary recursion tree**, making this a classic example of:
- Backtracking
- Multiple (binary) recursion

## âœ… Solution: Generate All Subsequences


````
def subsequence(s, i, curr, result):
    # Base case: all characters processed
    if i == len(s):
        result.append(curr)
        return

    # Include current character
    subsequence(s, i + 1, curr + s[i], result)

    # Exclude current character
    subsequence(s, i + 1, curr, result)


path = []
subsequence('abc', 0, '', path)
path
````

## â± Time and Space Complexity

- **Time Complexity:** `O(2^n)`  
- **Space Complexity:** `O(n)` (recursion stack)
