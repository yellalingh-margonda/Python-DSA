
# ðŸ” Subsequence Sum vs Combination Sum I & II  
## (Why binary choice vs N-ary choice matters)

This notebook fixes a **very important conceptual point**:

> ðŸ”‘ **Subsequence problems are BINARY choice problems**  
> ðŸ”‘ **Combination problems are N-ARY choice problems**

This single distinction explains:
- why pick / not-pick works for subsequence
- why for-loop recursion is required for combination sum



## ðŸ§  Core Concept: How Many Choices at Each Level?

### 1ï¸âƒ£ Subsequence Sum â†’ **2 choices**
At every index:
- Pick the element
- Not pick the element

âž¡ï¸ Binary recursion tree (2 branches)

---

### 2ï¸âƒ£ Combination Sum â†’ **N choices**
At every recursion level:
- You can choose **any element from index â†’ n**
- Each choice starts a new branch

âž¡ï¸ N-ary recursion tree (loop-based)

This is the MOST important difference.



## âœ… Subsequence Sum  
### Why Pick / Not Pick WORKS

Because:
- Every element is **independent**
- Each element contributes **two decisions**
- We move index forward in both cases

### Mental Model
> â€œFor this element, do I take it or skip it?â€


```

def subsequence_sum(arr, index, target, path):
    if index == len(arr):
        if target == 0:
            return [path[:]]
        return []

    # PICK
    path.append(arr[index])
    pick = subsequence_sum(arr, index + 1, target - arr[index], path)
    path.pop()

    # NOT PICK
    not_pick = subsequence_sum(arr, index + 1, target, path)

    return pick + not_pick

print(subsequence_sum([1,2,1], 0, 2, []))
```


## âŒ Why Pick / Not Pick FAILS for Combination Sum

Pick / not-pick creates **only 2 branches**, but:

### Combination Sum needs:
- Trying **every possible next candidate**
- Order-independent selection
- Duplicate control
- Optional reuse of elements

âž¡ï¸ These require **more than 2 choices per recursion**.

Hence:
> âŒ Binary recursion is insufficient  
> âœ… Loop-based (N choices) recursion is required



## âœ… Combination Sum I (Reuse Allowed)  
### FOR-LOOP Version (Correct & Preferred)

### Key Rules
- Each recursion level chooses **one of many candidates**
- Reuse allowed â†’ recursive call uses same index
- Loop represents **N choices**

### Mental Model
> â€œFrom here, which number do I choose next?â€


```

def combination_sum_1(arr, target):
    res = []

    def backtrack(start, target, path):
        if target == 0:
            res.append(path[:])
            return

        for i in range(start, len(arr)):
            if arr[i] > target:
                continue
            path.append(arr[i])
            backtrack(i, target - arr[i], path)  # reuse allowed
            path.pop()

    backtrack(0, target, [])
    return res

print(combination_sum_1([2,3,6,7], 7))
```


## âœ… Combination Sum II (No Reuse, No Duplicates)

### Why Loop + Duplicate Skip?
- Each level still has **N choices**
- But:
  - Elements used once
  - Duplicate combinations forbidden

### Mental Model
> â€œChoose one unused candidate, then move forward.â€


```

def combination_sum_2(arr, target):
    arr.sort()
    res = []

    def backtrack(start, target, path):
        if target == 0:
            res.append(path[:])
            return

        for i in range(start, len(arr)):
            if i > start and arr[i] == arr[i-1]:
                continue
            if arr[i] > target:
                break
            path.append(arr[i])
            backtrack(i + 1, target - arr[i], path)
            path.pop()

    backtrack(0, target, [])
    return res

print(combination_sum_2([10,1,2,7,6,1,5], 8))
```


## ðŸ” Side-by-Side Mental Comparison

| Problem | Choices per level | Recursion type | Template |
|------|-----------------|---------------|----------|
| Subsequence Sum | 2 | Binary | Pick / Not Pick |
| Combination Sum I | N | N-ary | For loop + reuse |
| Combination Sum II | N | N-ary | For loop + skip dup |

## ðŸ§  FINAL GOLDEN RULE (Memorize)

> **Binary choice â†’ Pick / Not Pick â†’ Subsequence**  
> **Multiple choices â†’ For loop â†’ Combination**

If at a recursion level you can choose **one among many elements**,  
ðŸ‘‰ **Pick / not-pick is the WRONG template.**
