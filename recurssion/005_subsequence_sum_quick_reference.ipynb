# Subsequence Sum Patterns â€“ Quick Reference Notebook

This notebook is a **quick reference** for solving **subsequence / subset sum problems** using recursion and backtracking.

It covers **four common variants**:
1. Find **all** subsequences with sum = target
2. Find **any one** subsequence with sum = target
3. Find subsequence using a **boolean/flag approach**
4. **Count** subsequences with sum = target


## ðŸ§  Core Idea (Applies to ALL problems here)

- At every index, you have **two choices**:
  1. Include the element
  2. Exclude the element

- This creates **binary (multiple) recursion**
- Uses **backtracking** to manage the path

**Important distinction:**
- When *building paths* â†’ we pass `path`
- When *counting* â†’ we return numbers (no accumulator passed)


## 1ï¸âƒ£ Find ALL subsequences with sum = target

```
def sub_sum(array, target, path=None, current_total=0, index=0):
    if path is None:
        path = []

    if current_total == target:
        return [path[:]]

    if index >= len(array) or current_total > target:
        return []

    path.append(array[index])
    include = sub_sum(array, target, path, current_total + array[index], index + 1)

    path.pop()
    exclude = sub_sum(array, target, path, current_total, index + 1)

    return include + exclude

print(sub_sum([1, 2, 1], 2))
```

### âœ” When to use this
- You must return **all valid subsequences**
- Output size matters


## 2ï¸âƒ£ Find ANY ONE subsequence with sum = target

```
def sub_single_sum(array, target, path=None, current_total=0, index=0):
    if path is None:
        path = []

    if current_total == target:
        return [path[:]]

    if index >= len(array) or current_total > target:
        return []

    path.append(array[index])
    include = sub_single_sum(array, target, path, current_total + array[index], index + 1)
    if include:
        return include

    path.pop()
    exclude = sub_single_sum(array, target, path, current_total, index + 1)
    if exclude:
        return exclude

    return []

print(sub_single_sum([1, 2, 1], 2))
```

### âœ” When to use this
- You only need **one valid subsequence**
- Early stopping improves performance


## 3ï¸âƒ£ Flag-based subsequence search

```
def subseq_sum_flag(array, target, index=0, path=None, current_sum=0):
    if path is None:
        path = []

    if current_sum == target:
        return path

    if index >= len(array) or current_sum > target:
        return False

    path.append(array[index])
    if subseq_sum_flag(array, target, index + 1, path, current_sum + array[index]):
        return path

    path.pop()
    if subseq_sum_flag(array, target, index + 1, path, current_sum):
        return path

    return None

print(subseq_sum_flag([3, 34, 4, 12, 5, 2], 9))
```

## 4ï¸âƒ£ Count subsequences with sum = target

```
def sub_seq_count(array, target, index=0, current_sum=0):
    if current_sum == target:
        return 1

    if index >= len(array) or current_sum > target:
        return 0

    count = 0
    count += sub_seq_count(array, target, index + 1, current_sum + array[index])
    count += sub_seq_count(array, target, index + 1, current_sum)

    return count

print(sub_seq_count([1, 2, 1], 2))
```

## ðŸ”‘ Key Observation

> When **counting**, we do **not** pass `count` as a parameter.
> Each recursive call returns its contribution.
